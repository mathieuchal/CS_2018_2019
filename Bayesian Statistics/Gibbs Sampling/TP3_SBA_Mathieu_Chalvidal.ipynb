### 1 - 2  Echantilloneur de Gibbs et par loi de probabibilité

```python
C1 = np.array([[1,0],[0,9]])
C2 = np.array([[5,4],[4,5]])
K = np.array([[5/9,-4/9],[-4,5]])
```

```python
def samples_norm_norm(C, n_samples=1000):
    samples_C = np.zeros((n_samples,2))
    for i in range(n_samples):
        samples_C[i] = multivariate_normal(np.zeros(2),C).rvs(1)
    return samples_C

samples_C1 = samples_norm_norm(C1, n_samples=5000)   
samples_C2 = samples_norm_norm(C2, n_samples=5000) 
```

```python
def Gibbs(Cov,sampler='deterministic',n_samples=1000):
    samples = np.zeros((n_samples,2))
    l=0
    C = np.linalg.inv(Cov)
    samples[0]= multivariate_normal(np.zeros(2),Cov).rvs(1)
    if sampler=='deterministic':
        for i in range(1,n_samples):
            l+=1
            if l%2==0:
                y=samples[i-1,1]
                x = norm((-(C[0,1])/(C[0,0]))*y,1/np.sqrt(C[0,0])).rvs(1)
                samples[i] = [x,y]
            else:
                x = samples[i-1,0]
                y = norm((-(C[0,1])/(C[1,1]))*x,1/np.sqrt(C[1,1])).rvs(1)
                samples[i] = [x,y]
    return samples
            
samples1 = Gibbs(C1, sampler='deterministic',n_samples=5000)      
samples2 = Gibbs(C2, sampler='deterministic',n_samples=5000) 
```

```python
fig, a = plt.subplots(1,2, figsize=(15,6))
a[0].scatter(samples1[:,0],samples1[:,1],s=0.5)
a[0].scatter(samples2[:,0],samples2[:,1],s=0.5)
a[1].scatter(samples_C[:,0],samples_C[:,1],s=0.5)
a[1].scatter(samples_C2[:,0],samples_C2[:,1],s=0.5)
```

**Observation** L'examen graphique semble montrer que les pour les 2 distributions proposées x1 et x2, l'échantilloneur de Gibbs donne des résultats similaires au tirage de la loi multidimensionelle normale connue. 

### 1 - 3 - Analyse de la convergence

Dans cette section, nous répétons l'opération d'échantillonage par méthode de Gibbs et par loi gaussienne multivariée pour des échantillons de taille $M = p \times 5000, p \in [1,10]$. Cela va nous permettre de comparer la convergence des estimateurs des moyennes de nos échantillons et leur variance en fonction du nb d'échantillon. 

```python
def Sampler(n_iter,n_samples,method='Gibbs'):
    samples = []
    Vars=[]
    if method=='Gibbs':
        Var=np.zeros((2,2))
        for i in range(n_iter):
            print('{}ème sampling'.format(i))
            samples.append(Gibbs(Cov=C1,sampler='deterministic',n_samples=n_samples))
            mean = np.mean(np.array(samples).flatten().reshape(2,-1),axis=1)
            Var += np.outer(mean,mean)
            Vars.append(np.trace(Var)/(i+1))
            
    elif method=='Loi':
        Var=np.zeros((2,2))
        for i in range(n_iter):
            print('{}ème sampling'.format(i))
            samples.append(samples_norm_norm(C1,n_samples=n_samples))
            mean = np.mean(np.array(samples).flatten().reshape(2,-1),axis=1)
            Var += np.outer(mean,mean)
            Vars.append(np.trace(Var)/(i+1))
        
    return samples,means,Vars

samp,means,Var = Sampler(10,5000,method='Gibbs')
samp1,means1,Var1 = Sampler(10,5000,method='Loi')
    
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,11),Var, label='Gibbs')
plt.plot(range(1,11),Var1, label='Loi')
plt.legend()
```

**Observation** : La variance des échantillons obtenus par la méthode de Gibbs affiche donc un comportement asymptotique plus lent que l'échantillonage par la loi directe. En revanche, dans des cas plus complexes ou dans le cas ou la loi de distribution multidimensionelle n'est pas connue, l'échantillonage de Gibbs peut s'avérer un outil pratique pour réaliser un échantillonage valable de la distribution d'intérêt. 

Ici pour améliorer l'échnatillonage de x2, nous pouvions simplement utiliser la relation linéaire obtenue à la question 1. En effet, pour une distribution gaussienne multivariée, une matrice de covariance diagonale est équivalente à l'indépendance des distributions. Ainsi l'échantillonage de Gibbs pour une loi de matrice de Covariance C1 se réduit à des tirages de la loi normale mono-dimensionelle. Il suffit ensuite d'appliquer l'opérateur linéaire K à l'échantillon tiré.

#### Explicitation du problème

D'après les hypothèses de l'énoncé, nous pouvons écrire **la vraissemblance de la variable y sachant x** comme

$$ p(y|x) =  \frac{1}{\sigma_b(2\pi)^{N/2}}e^{-\frac{1}{2\sigma_b^2}\|y-Ax\|^{2}}$$

La **loi à priori de $x,\tau$** s'écrit par la formule de Bayes:

$$ p(x, \tau) = p(x|\tau)p(\tau) = \frac{\tau^{M/2}}{(2\pi)^{M/2}}e^{-\frac{\tau\| x\|^{2}}{2}} . \frac{\tau^{\alpha-1} \beta^{\alpha} e^{-\beta\tau}}{\Gamma(\alpha)} $$

Cette loi peut se réecrire de manière plus compacte comme: 

$$ p(x, \tau) =  p(x|\tau)p(\tau) = \frac{\tau^{\alpha-1 + M/2}\beta^{\alpha}}{\Gamma(\alpha)(2\pi)^{M/2}}e^{-\tau(\sum \frac{x^{2}}{2}-\beta)} \sim \Gamma(\alpha + \frac{M}{2},\sum \frac{x^{2}}{2}+\beta)$$

L'avantage de l'introduction d'une loi Gamma pour l'a-priori de tau est donc **le calcul de la loi conjuguée de $x,\tau$ comme une loi Gamma** de paramètre $(\alpha-1 + \frac{M}{2},\sum \frac{x^{2}}{2}-\beta)$ et qui est facilement échantillonable.

La **loi à posteriori de $x,\tau$** sachant y vaut alors, utilisant à nouveau la formule de Bayes:

$$ p(x, \tau |y) = \frac{p(x, \tau ,y)}{p(y)} = \frac{p(y|x, \tau)p(x, \tau)}{p(y)} = \frac{p(y|x)p(x|\tau)p(\tau)}{p(y)}  $$

Enfin, en utilisant les résultats précédants, il vient: 

$$ p(x, \tau |y) \propto \frac{1}{\sigma_b(2\pi)^{N/2}}e^{-\frac{1}{2\sigma_b^2}\|y-Ax\|^{2}}.\frac{\tau^{\alpha-1 + M/2}\beta^{\alpha}}{\Gamma(\alpha)(2\pi)^{M/2}}e^{-\tau(\sum \frac{x^{2}}{2}-\beta)} $$

### 2 -1 Esperance a posteriori par echantillonage de Gibbs

**Echantillonage de Gibbs**:
Tirage alternatif suivant les lois $p(x|y,\tau)$ and  $p(\tau|y,x)$. Dans notre cas, A est la matrice de Toeplitz correspondante à la convolution du signal discret.

Pour $x \in R^{N}$
> $$ p(x |\tau,y) = \frac{p(x, \tau,y)}{p(\tau,y)} \propto p(y|x)p(x|\tau) \sim N(\frac{y}{1+\sigma_b^2\tau},\frac{I_N}{\frac{1}{\sigma_b^2}+\tau}) $$

Si l'on fait l'hypothèse que $\sigma_b \approx \sigma_x$, il vient: $\sim N(\frac{y}{4},\frac{I_N}{2\tau})$

Pour $\tau \in R$
> $$ p(\tau |x,y) = p(x|\tau)p(\tau) \sim \Gamma(\alpha + \frac{M}{2},\sum \frac{x^{2}}{2}+\beta)$$


Et finalement nous avons l'estimateur a posteriori suivant:
$$\hat x_{EAP} = \int\limits_{R^{N}} x p(x|\tau,y)dx \approx \frac{1}{n}\sum\limits_{i=1}^{n} x_i$$

en utilisant la loi des grands nombres et où $x_i$ est tiré selon la loi $p(x|\tau,y)$

```python
padding1 = np.zeros(118, h.dtype)
padding2 = np.zeros(118, h.dtype)
first_col = np.r_[h, padding1]
first_row = np.r_[h[0], padding2]
A = scipy.linalg.toeplitz(first_col, first_row)

def Gibbs(signal, alpha, beta, sampler='deterministic', n_samples=1000):
    N = signal.shape[0]
    x_samples = np.zeros((n_samples,N))
    tau_samples = np.zeros(n_samples)
    l=0
    if sampler=='deterministic':
        tau_samples[0] = np.random.gamma(alpha,beta,1)
        for i in range(1,n_samples):
            if i%1000==0:
                print(i)
            tau = tau_samples[i-1]
            
            #Cov = np.eye(N)/(1/sigma_b**2+tau)
            #x_samples[i] = multivariate_normal(signal/(1+sigma_b**2*tau),Cov).rvs(1)
            
            Cov = np.eye(N)/(2*tau)
            x_samples[i] = multivariate_normal(signal/4,Cov).rvs(1)
            
            tau_samples[i] = np.random.gamma(alpha + N/2, beta + 0.5*np.linalg.norm(signal - (A@x_samples[i-1])[10:-10],2))
            
    x_EAP = np.mean(x_samples, axis=0)        
    return x_EAP, x_samples, tau_samples
 
```

### 2-2 Maximum de vraissemblance

L'estimation du maximum de vraissemblance peut être calculé au moyen de la loi jointe de vraissemblance présentée plus haut: 

$$ p(x, \tau |y) = \frac{1}{\sigma_b(2\pi)^{N/2}}e^{-\frac{1}{2\sigma_b^2}\|y-Ax\|^{2}}.\frac{\tau^{\alpha-1 + M/2}\beta^{\alpha}}{\Gamma(\alpha)(2\pi)^{M/2}}e^{-\tau(\sum \frac{x^{2}}{2}-\beta)} $$

La densité du produit des lois normales est $C^{\infty}$, nous pouvons alors écrire les conditions du premier ordre pour l'estimation de $\hat x_{MV}$ et $\hat \tau_{MV}$ par log-vraissemblance:

$$ \nabla_x log(p(x, \tau |y)) = 0_{R^N} \text{and},   \frac{\partial log(p(x, \tau |y))}{\partial \tau} =0$$

Ceci nous donne finalement; 

> $$\hat x_{EAP} = (A^tA + Id)^{-1}A^ty$$

> $$ \hat \tau_{EAP}  = \frac{\alpha -1 + M/2}{\frac{\|\hat x_{EAP}\|^2}{2}-\beta}$$

```python
padding1 = np.zeros(98, h.dtype)
padding2 = np.zeros(118, h.dtype)
first_col = np.r_[h, padding1]
first_row = np.r_[h[0], padding2]
A2 = scipy.linalg.toeplitz(first_col, first_row)
M = m1.shape[0]
alpha=2
beta=1

x_MV1 = np.linalg.inv(A2.T@A2 + np.eye(M))@(A2.T@m1)
x_MV2 = np.linalg.inv(A2.T@A2 + np.eye(M))@(A2.T@m2)
x_MV3 = np.linalg.inv(A2.T@A2 + np.eye(M))@(A2.T@m3)

Tau_MV1 = (alpha-1+M/2)/(0.5*np.linalg.norm(x_MV1)**2 - beta)
Tau_MV2 = (alpha-1+M/2)/(0.5*np.linalg.norm(x_MV2)**2 - beta)
Tau_MV3 = (alpha-1+M/2)/(0.5*np.linalg.norm(x_MV3)**2 - beta)

print('Tau_MV, signal 1 = {}'.format(Tau_MV1))
print('Tau_MV, signal 2 = {}'.format(Tau_MV2))
print('Tau_MV, signal 3 = {}'.format(Tau_MV3))

plt.plot(A2@x_MV1)
plt.plot(A2@x_MV2)
plt.plot(A2@x_MV3)
```
